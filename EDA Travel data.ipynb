# Method 1: Creating binary variables through one hot encoding


```python
a_encoded = pd.get_dummies(a,drop_first=True)
```

```python
a_encoded
```

```python
# by using sklearn
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder(categories="auto", drop= "first", sparse=False)
sk_encoder = encoder.fit(a)
sk_encoder
```

```python
a_transformed = encoder.transform(a)
a_transformed
```

# Method 2: Creating binary variables through ordinal numbers

```python
# by using sklearn
from sklearn.preprocessing import OrdinalEncoder
```

```python
enc = OrdinalEncoder()
enc.fit(a)
a_trans = enc.transform(a)
a_trans
```
